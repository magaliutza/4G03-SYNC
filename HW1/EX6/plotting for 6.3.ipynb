```python
import numpy as np
import matplotlib.pyplot as plt
import math

incoming = np.genfromtxt('dat.csv', delimiter = ',', skip_header = 1, usecols = (0,1,2,3), dtype=None)

#print(incoming[5][1])
trap_inter = [incoming[i][1] for i in range(0,15)]
trap_inter_2 = [math.log(i,2) for i in trap_inter]
trap_timer = [incoming[i][2] for i in range(0,15)]
trap_preciser = [incoming[i][3] for i in range(0,15)]

n = 7

trap_inter_z = [incoming[i][1] for i in range(0,7)]
trap_inter_2_z = [math.log(i,2) for i in trap_inter]
trap_timer_z = [incoming[i][2] for i in range(0,7)]
trap_preciser_z = [incoming[i][3] for i in range(0,7)]




#print(incoming)
#print(trap_timer)

romb_inter = [incoming[i][1] for i in range (15,n+15)]
romb_timer = [incoming[i][2] for i in range (15,n+15)]
romb_preciser = [incoming[i][3] for i in range (15,n+15)]

# trap
plt.grid()
plt.scatter(trap_inter_z, trap_timer_z, c='blue', marker='o', label = 'test')
plt.xlabel("Number of steps 'n' called in the trapezoid func.")
plt.ylabel("Time to compute 50,000 approximations, in seconds")

plt.savefig("traptime", dpi=200)
plt.show()

plt.scatter(trap_inter_z, trap_preciser_z, c='blue', marker='o')
plt.grid()
plt.xlabel("Number of steps 'n' called in the trapezoid func.")
plt.ylabel("Precision of result, as a percent.")

plt.savefig("trapprec", dpi=200)
plt.show()

#romb

plt.grid()
plt.scatter(romb_inter, romb_timer, c='green', marker='o', label = 'test')
plt.xlabel("Order 'n' approximation called in the Romberg R(n,n).")
plt.ylabel("Time to compute 50,000 approximations, in seconds")

plt.savefig("rombtime", dpi=200)
plt.show()

plt.grid()
plt.scatter(romb_inter, romb_preciser, c='green', marker='o', label = 'test')
plt.xlabel("Order 'n' approximation called in the Romberg R(n,n).")
plt.ylabel("Time to compute 50,000 approximations, in seconds")

plt.savefig("rombprec", dpi=200)
plt.show()
```
