```python
import matplotlib.pyplot as plt
import numpy as np
import math

input = np.genfromtxt('energies.csv', delimiter = ',', skip_header = 1)
    
e0known = -4.5154463544
m = [input[i][0] for i in range(len(input))]
e0 = [input[i][1] for i in range(len(input))]
e1 = [input[i][2] for i in range(len(input))]

diff = [abs(e0-e0known) for e0 in e0]

plt.plot(m, e0, color='silver', marker = 'o', label = "E_0")
plt.plot(m, diff, color='gold', marker = 'o', label = "|E_0-E_exact|")
plt.plot(m, e1, color='red', marker = 'o', label = "E_1")
plt.xlabel('M')
plt.ylabel('Energy')
plt.title('Energy vs M for L=10')
plt.legend()
plt.grid(True)
plt.savefig('energy10.png', dpi = 200)
plt.show()

plt.plot(m, e0, color='silver', marker = 'o', label = "E_0")
plt.plot(m, diff, color='gold', marker = 'o', label = "|E_0-E_exact|")
plt.plot(m, e1, color='red', marker = 'o', label = "E_1")
plt.xlabel('M')
plt.ylabel('Energy')
plt.title('Energy vs M for L=10')
plt.legend()
plt.xlim(4,16)
plt.ylim(-.5,1)
plt.grid(True)
plt.savefig('energy10zoom.png', dpi = 200)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

input3 = np.genfromtxt('energies_2.csv', delimiter = ',', skip_header = 1)
    
L = [input3[i][0] for i in range(len(input3))]
e0 = [input3[i][1] for i in range(len(input3))]
e1 = [input3[i][2] for i in range(len(input3))]

diff_2 = [abs(e1-e0) for e1,e0 in zip(e1,e0)]

plt.plot(L, diff_2, color='gold', marker = 'o', label = "|E_1-E_0|")
plt.xlabel('L')
plt.ylabel('Energy Diffrence')
plt.title('Energy Diffrence |E_1-E_0| vs even L values for M=50')
plt.legend()
plt.grid(True)
plt.savefig('energy_2.png', dpi = 200)
plt.show()


plt.figure(figsize=(10, 5))
plt.plot(L, diff_2, color='gold', marker = 'o', label = "|E_1-E_0|")
plt.xlabel('log(L)')
plt.ylabel('log(Energy Diffrence)')
plt.yscale('log')
plt.xscale('log')
plt.title('Energy Diffrence log|E_1-E_0| vs even log(L) values for M=50')
plt.xlim(10,)
plt.legend()
plt.grid(True)
plt.savefig('energy_2_log.png', dpi = 200)
plt.show()

```
