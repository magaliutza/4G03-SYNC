```python
import matplotlib.pyplot as plt
import numpy as np
import math

input = np.genfromtxt('ge.txt', delimiter = ',', skip_header = 1)
    
e = [input[i][0] for i in range(len(input))]
g = [input[i][1] for i in range(len(input))]

plt.plot(e, g, color='gold', marker = 'o', markersize = 3)
plt.xlabel('E')
plt.ylabel('G(E)')
plt.title('G(E) vs E for xm = 5')
#plt.legend()
#plt.axvline(47.8651)
plt.ylim(-1,1)
#plt.xlim(34.075,34.11)
plt.grid(True)
plt.savefig('g.png', dpi = 200)
plt.show()

```

```python
input2 = np.genfromtxt('eigenfunc.txt', delimiter = ',', skip_header = 1)
    
x = [input2[i][0] for i in range(len(input2))]
phil = [input2[i][1] for i in range(len(input2))]
phir = [input2[i][2] for i in range(len(input2))]

turning = 5.0
index_turn = min(range(len(x)), key=lambda i: abs(x[i] - turning))

x_left = x[:index_turn + 1]
phil = phil[:index_turn + 1]
#phil = [phil*phil for phil in phil]
x_right = x[index_turn:]
phir = phir[index_turn:]
#phir = [phir*phir for phir in phir]

plt.plot(x_left, phil, color='blue', marker = 'o')
plt.plot(x_right, phir, color='blue', marker = 'o')


plt.xlabel('x')
plt.ylabel('Wavefunc $\phi$')
plt.title('Wavefunction at Eigenval = 47.8691 ')
#plt.legend()
#plt.ylim(-10,10)
#plt.xlim(-14,-5)|
plt.grid(True)
plt.savefig('eig_47.png', dpi = 200)
plt.show()
```
