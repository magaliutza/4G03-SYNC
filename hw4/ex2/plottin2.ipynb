```python
import matplotlib.pyplot as plt
import numpy as np
import math

input = np.genfromtxt('Umid.txt', delimiter = ',', skip_header = 0)
    
x05 = input[:,49]
x = np.linspace(0,1,100)

plt.plot(x, x05, color='gold', marker = 'o', markersize = 0.2)
plt.xlabel('x')
plt.ylabel('U')
plt.title('U(x=1/2, y)')
#plt.legend()
#plt.axvline(47.8651)
#plt.ylim(-1,1)
#plt.xlim(34.075,34.11)
plt.grid(True)
plt.savefig('ux12.png', dpi = 200)
plt.show()

plt.xlabel('x')
plt.ylabel('y')
plt.title("Heat map of U(x,y)")
zmax = np.max(input)
zmin = np.min(input)
plt.contourf(x, x, input, 51, cmap=plt.cm.gnuplot2, vmax=zmax, vmin=zmin) 
plt.savefig('heat.png', dpi = 200)
plt.show()


input2 = np.genfromtxt('Utheo.txt', delimiter = ',', skip_header = 0)

theo05 = input2[:,49]
plt.plot(x, theo05, color='gold', marker = 'o', markersize = 0.2)
plt.xlabel('x')
plt.ylabel('U')
plt.title('Theoretical U(x=1/2, y)')
#plt.legend()
#plt.axvline(47.8651)
#plt.ylim(-1,1)
#plt.xlim(34.075,34.11)
plt.grid(True)
plt.savefig('theo05.png', dpi = 200)
plt.show()

theoL = input2[-1, :]
plt.plot(x, theoL, color='gold', marker = 'o', markersize = 0.2)
plt.xlabel('y')
plt.ylabel('U')
plt.title('Theoretical U(x, y = L)')
#plt.legend()
#plt.axvline(47.8651)
#plt.ylim(-1,1)
#plt.xlim(34.075,34.11)
plt.grid(True)
plt.savefig('theoL.png', dpi = 200)
plt.show()
```
